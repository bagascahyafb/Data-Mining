```python
import pandas as  pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.preprocessing import Normalizer
```

# Menampilkan dataset

```python
titanic_df = pd.read_csv('titanic.csv')
titanic_df
```

# Menampilkan jumlah baris dan kolom pada dataset

```python
titanic_df.shape
```

# Mengambil kolom Age, Fare sebagai data

```python
data = titanic_df[['Age', 'Fare']]
data
```

# Mengambil kolom Survived sebagai class

```python
Class = titanic_df['Survived']
Class
```

# mengisi missing value kolom Age berdasarkan classnya

```python
df = pd.concat([data, Class], axis=1)
df = df.groupby('Survived')[['Age', 'Fare']].transform(lambda x: x.fillna(x.mean()))
data = df
data
```

# Melakukan Normalisasi data dengan beberapa metode

## MinMax Scaler

```python
scaler = MinMaxScaler()
minmax = pd.DataFrame(scaler.fit_transform(data), index=data.index, columns=data.columns)
minmax
```

## Z-Score

```python
scaler = StandardScaler()
zscore = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)
zscore
```

## Sigmoidal

```python
sigmoid = 1 / (1+np.exp(-data))
sigmoid
```

```python
sigmoid_scaler = Normalizer()
sigmoid = pd.DataFrame(sigmoid_scaler.fit_transform(data), index=data.index, columns=data.columns)
sigmoid
```
