```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
titanic_df = pd.read_csv('titanic.csv')
titanic_df.head(10)
```

```python
titanic_df.shape
```

```python
titanic_df_id = titanic_df[['Name', 'Sex', 'Age', 'Pclass', 'Fare']]
titanic_df_id
```

```python
titanic_class = titanic_df['Survived']
titanic_class
```

```python
titanic_df['fitur'] = titanic_df['SibSp'] + titanic_df['Parch']
titanic_df
```

```python
jumlah_penumpang = titanic_df['Pclass'].value_counts()

print('Jumlah Penumpang pada setiap class :\n', jumlah_penumpang)
```

```python
jumlah_penumpang_gender = titanic_df['Sex'].value_counts()

print('Jumlah Penumpang untuk Setiap Jenis Kelamin :\n', jumlah_penumpang_gender)
```

```python
jumlah_penumpang_selamat = titanic_df.groupby(['Pclass', 'Survived']).size()

print('Jumlah Penumpang selamat berdasarkan class :\n',jumlah_penumpang_selamat)
print('0 = Tidak Selamat\n1 = Selamat')
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=titanic_df, x=titanic_df.index, y='Sex', hue='Pclass', palette='Set1')
plt.xlabel('Urutan Data')
plt.ylabel('Sex')
plt.legend(title='Pclass', loc='upper right')
plt.title('Visualisasi Jumlah Penumpang untuk Setiap Pclass')
plt.show()
```

```python
titanic_df = titanic_df.dropna(subset=['Age'])
plt.figure(figsize=(10,6))
sns.scatterplot(data=titanic_df, x=titanic_df.index, y='Age', hue='Surv', palette='Set1')
plt.xlabel('Urutan Data')
plt.ylabel('Age')
plt.legend(title='Pclass', loc='upper right')
plt.title('Visualisasi Umur Penumpang untuk Setiap Pclass')
plt.show()
```
