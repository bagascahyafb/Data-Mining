```python
print("Data Valid - Before :", west_df['BEDROOMS'].count())
missingRows = west_df.sample(10).index
west_df.loc[missingRows, ['BEDROOMS']] = np.nan

print("Data Valid - After NaN :", west_df['BEDROOMS'].count())
west_df["BEDROOMS"].count()
```

```python
reduce_df = west_df.dropna()
print("Data Valid - After DROP NaN :", len(reduce_df))
```

```python
medianBedroom = west_df["BEDROOMS"].median()
print("Data Valid - Nilai Median :", medianBedroom)

west_df.BEDROOMS = west_df.BEDROOMS.fillna(value=medianBedroom)
print("Data Valid - After Median :", west_df['BEDROOMS'].count())
```

## Data Visualisation

```python
import os
import numpy as np
import networkx as nx
import pandas as pd
from pandas.plotting import scatter_matrix, parallel_coordinates
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
boston_df = pd.read_csv("BostonHousing.csv")
boston_df = boston_df.rename(columns= {"CAT. MEDV" : 'CAT_MEDV'})
boston_df.head(10)
```

```python
boston_df.plot.scatter(x = 'LSTAT', y = 'MEDV', legend = False)
plt.show
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(5, 3)

ax.scatter(boston_df.LSTAT, boston_df.MEDV, color='C2', facecolor='None')
plt.xlabel('LSTAT')
plt.ylabel('MEDV')

plt.tight_layout()
plt.show()
```

```python
ax = boston_df.groupby('CHAS').mean().MEDV.plot(kind='bar', figsize=[5,3])
ax.set_ylabel("AVG MEDV")
plt.tight_layout()
plt.show()
```

```python
dataplot = boston_df.groupby('CHAS').mean().MEDV
fig, ax = plt.subplots()
ax.bar(dataplot.index, dataplot, color=['C5', 'C1'])
ax.set_xticks([0,1])

ax.set_xlabel("CHAS")
ax.set_ylabel("AVG MEDV")

plt.show()
```

### Ilustrasi visualisasi jaringan

```python
ebay_df = pd.read_csv('eBayNetwork.csv')

G = nx.from_pandas_edgelist(ebay_df, source='Seller', target='Bidder')

isibidder = [i in set(ebay_df.Bidder) for i in G.nodes()]
pos = nx.spring_layout(G, k = 0.13, iterations=60, scale=0.5)

plt.figure(figsize=[10,10])
nx.draw_networkx(G, pos=pos, with_labels=False, edge_color='lightgray', 
node_color=['gray' if bidder else 'black' for bidder in isibidder], 
node_size=[50 if bidder else 200 for bidder in isibidder])

plt.axis('off')
plt.show()
```
