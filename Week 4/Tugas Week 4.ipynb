```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

# Menampilkan Dataset

```python
# Train
dataset = pd.read_csv('titanic.csv')
print("Train Dataset:")
dataset.head(10)
```

```python
# Test
test_dataset = pd.read_csv('titanic_test.csv')
print("Test Dataset:")
test_dataset
```

# Membuang missing value pada data train dan test

```python
# Train
train_data = dataset[['Age', 'Fare']].dropna()
print(train_data)
pos_missing_train = dataset.index.difference(train_data.index)
print("Banyaknya Missing Value: ", len(pos_missing_train))
```

```python
# Test
test_data = test_dataset[['Age', 'Fare']].dropna()
print(test_data)
pos_missing_test = test_dataset.index.difference(test_data.index)
print("Banyaknya Missing Value: ", len(pos_missing_test))
```

# Mengambil class label pada data train dan test

```python
# Train
train_label = dataset.loc[train_data.index, 'Survived']
print("Train Label:")
print(train_label)
```

```python
# Test
test_label = pd.read_csv('titanic_testlabel.csv').loc[test_data.index]
test_label = test_label['Survived']
print("Test Label:")
print(test_label) 
```

# Melakukan Normalisasi pada data train dan test

```python
scaler = MinMaxScaler()
# train
train_data = pd.DataFrame(scaler.fit_transform(train_data), columns=train_data.columns)
print("Train Data:")
print(train_data)
```

```python
# Test
test_data = pd.DataFrame(scaler.transform(test_data), columns=test_data.columns)
print("Test Data:")
print(test_data)
```

# Melakukan Klasifikasi K-NN pada data dan cek eror rationya

```python
for k in range(1, 11):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(train_data, train_label)
    class_result = knn.predict(test_data)
    error_ratio = 1 - accuracy_score(test_label, class_result)
    print(f"Error ratio for k={k}: {error_ratio}")
```
